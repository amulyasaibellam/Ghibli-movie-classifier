```python
from duckduckgo_search import ddg_images
from fastcore.all import *
from fastdownload import download_url

def search_images(term, max_images=30):
    print(f"Searching for '{term}'")
    return L(ddg_images(term, max_results=max_images)).itemgot('image')
```

```python
path = Path('./new_data')
from time import sleep
from fastai.vision.all import *

movie_names = ['Princess Mononoke',
               'Spirited Away',
               'Howl\'s Moving Castle',
               'My Neighbor Totoro',
               'Kiki\'s Delivery Service',
               'The Tale of Princess Kaguya',
               'Nausicaa of the Valley of the Wind',
               'Castle in the Sky',
               'Grave of the Fireflies',
               'Whisper of the Heart']

folders_and_characters = {'Princess_Mononoke' : ['Ashitaka',
                                                 'San',
                                                 'Moro',
                                                 'Lady Eboshi',
                                                 'Deer God',
                                                 'Forest spirits',
                                                 'Okkoto'],
                          'Spirited_away': ['Chihiro',
                                            'Kaonashi',
                                            'Haku',
                                            'Yubaba',
                                            'Aogeru',
                                            'Boh'],
                          'Howls_moving_castle': ['Wizard Howl',
                                                  'Sophie Hatter',
                                                  'Calcifer',
                                                  'Markl',
                                                  'Witch of the waste',
                                                  'Kakashi no Kabu'],
                          'My_Neighbor_Totoro': ['Catbus',
                                                 'Mei',
                                                 'Satsuki',
                                                 'totoro',
                                                 'Kanta'],
                          'Kikis_delivery_service': ['Kiki',
                                                     'Tombo',
                                                     'Jiji',
                                                     'Ursula',
                                                     'Osono',
                                                     'Bakery Worker'],
                          'Princess_Kaguya': ['Princess Kaguya',
                                              'Sutemaru',
                                              'Menowarawa',
                                              'Celestial beings',
                                              'Ishitsukuri'],
                          'Nausicaa': ['Nausicaa',
                                       'Asbel',
                                       'Kushana',
                                       'Yupa',
                                       'Teto'],
                          'Castle_in_the_Sky': ['Sheeta',
                                                'Pazu',
                                                'Fox squirrel',
                                                'Muska',
                                                'Dola',
                                                'Motro'],
                          'Grave_of_the_Fireflies': ['Seita',
                                                     'Setsuko',
                                                     'Seita\'s aunt',
                                                     'Mrs Yakokawa',
                                                     'Kiyoshi Yakokawa'],
                          'Whisper_of_the_heart': ['Seiji',
                                                   'Shizuku',
                                                   'The Baron',
                                                   'Sugimura',
                                                   'Shiro Nishi']}

for (movie_name, folder) in zip(movie_names, folders_and_characters.keys()):
    dest = (path/folder)
    dest.mkdir(exist_ok=True, parents=True)
    download_images(dest, urls=search_images(movie_name + ' movie characters', 50))
    sleep(10)
    for charac in folders_and_characters.get(folder):
        download_images(dest, urls=search_images(charac + ' from ' + movie_name, 10))
        sleep(10)
    resize_images(path/folder, max_size=400, dest=path/folder)
```
